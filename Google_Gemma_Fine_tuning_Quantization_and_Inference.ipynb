## Quantization of Gemma 7B

```python
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    set_seed,
    BitsAndBytesConfig
)

model_id: str = "google/gemma-7b"
set_seed(1234)
tokenizer = AutoTokenizer.from_pretrained(model_id)

bnb_config = BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_compute_dtype = torch.float16,
    bnb_4bit_quant_type = "fp4",
    bnb_4bit_use_double_quant = True
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    trust_remote_code=True,
    device_map="cuda")

prompt: str = "The best recipe for pasta is"

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, do_sample=True, max_new_tokens=100)
tokenizer.decode(outputs[0], skip_special_tokens=True)
```

## Fine-tuning Gemma 7B with QLORA

```python
!pip install -qU datasets trl unsloth
```

```python
from unsloth import is_bfloat16_supported
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    set_seed,
    BitsAndBytesConfig
)
```

```python
model_id: str = "google/gemma-7b"

set_seed(1234)

tokenizer = AutoTokenizer.from_pretrained(
    model_id,
    add_eos_token=True,
    use_fast=True
)

tokenizer.pad_token = tokenizer.eos_token
tokenizer.pad_token_id = tokenizer.eos_token_id
tokenizer.padding_side="left"
```

```python
from datasets import load_dataset, load_dataset_builder

dataset_name: str = "timdettmers/openassistant-guanaco"

dataset_builder = load_dataset_builder(dataset_name)
print(dataset_builder.info.features)
print(dataset_builder.info.splits)
```

```python
from peft import prepare_model_for_kbit_training, LoraConfig
from trl import SFTTrainer, SFTConfig

dataset = load_dataset(dataset_name, trust_remote_code=True)

bnb_config = BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_compute_dtype = torch.float16,
    bnb_4bit_quant_type = "nf4",
    bnb_4bit_use_double_quant = True
)

pretrained_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    trust_remote_code=True,
    device_map={"": 0}
)
```

```python
model = prepare_model_for_kbit_training(pretrained_model)

# Configure the pad token in the model
model.config.pad_token_id = tokenizer.pad_token_id
model.config.use_cache = False

peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.05,
    r=16,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules= ['k_proj', 'q_proj', 'v_proj', 'o_proj', "gate_proj", "down_proj", "up_proj"]
)

sft_arguments = SFTConfig(
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    warmup_steps = 5,
    do_eval = True,
    num_train_epochs = 3, # Set this for 1 full training run.
    max_steps = 300,
    eval_strategy="steps",
    eval_steps = 50,
    logging_steps=50,
    learning_rate = 2e-4,
    fp16 = True,
    bf16 = False,
    optim = "adamw_8bit",
    weight_decay = 0.01,
    lr_scheduler_type = "linear",
    seed = 3407,
    output_dir = "model_traning_outputs",
    report_to = "none",
    max_seq_length = 2048,
    dataset_num_proc = 4,
    packing = False, # Can make training 5x faster for short sequences.
)

model.add_adapter(peft_config, adapter_name="lora_2")

device = torch.device("cuda")

trainer = SFTTrainer(
    model=model.to(device),
    train_dataset=dataset['train'],
    eval_dataset=dataset['test'],
    peft_config=peft_config,
    tokenizer=tokenizer,
    args=sft_arguments
)

trainer.train()
```
